```python
# Import relevant modules
import pandas as pd
import numpy as np
from dynamic import Submarine, ClosedLoop
from control import PDController  # Import your PDController class
from dynamic import Mission  # Import your Mission class

# Instantiate the submarine
sub = Submarine()

# Instantiate your PD controller with specified gains
kp = 0.4
kd = 0
controller = PDController(kp, kd)

# Load mission data from CSV
mission = Mission.from_csv("mission.csv")  # Adjust the path to your CSV file


# Create the closed loop system
closed_loop = ClosedLoop(sub, controller)

# Run the simulation
trajectory = closed_loop.simulate_with_random_disturbances(mission)

# Plot the completed mission
trajectory.plot_completed_mission(mission)
```
