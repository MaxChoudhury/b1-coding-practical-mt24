# Simulate closed-loop
After implementing your the controller, this next section will allow it to run and produce a plot.

```python
# Import relevant modules
import pandas as pd
import numpy as np
from uuv_mission.dynamic import Submarine, ClosedLoop, Mission # Import your Submarine, ClosedLoop, and Mission classes
from uuv_mission.control import PDController  # Import your PDController class

# Instantiate the submarine
sub = Submarine()

# Instantiate your PD controller with specified gains
kp = 0.15
kd = 0.6
controller = PDController(kp, kd)

# Load mission data from CSV
mission = Mission.from_csv(r'C:\Users\max_c\OneDrive\Desktop\University\B1\B1 Scientific Computation\Version3\b1-coding-practical-mt24\uuv_mission\data\mission.csv')  # Adjust the path to your CSV file


# Create the closed loop system
closed_loop = ClosedLoop(sub, controller)   

# Run the simulation
trajectory = closed_loop.simulate_with_random_disturbances(mission)

# Plot the completed mission
trajectory.plot_completed_mission(mission)
```
